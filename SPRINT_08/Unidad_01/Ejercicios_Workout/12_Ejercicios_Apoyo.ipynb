
### Ejercicio 3

Completa el cdigo para poder reproducir la imagen de la figura, tendrs que revisar el parmetro "connectionstyle" del rango de parmetros del argumento "arrowprops". 
Nota: Las coordenadas del texo no estn dadas como puntos de la grfica sino como "offset" del punto sealado. Esto quiere decir que si el offset es la tupla (-2,10) y las coordenadas (xy) del punto sealado por la anotacin son (30,40), por ejemplo, el texto se situar en (30-2, 40 + 10) es decir en (28,50) [Recuerda, es la coordenada de la esquina inferior izquierda del cuadro de texto "invisible" que contiene al texto]

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(8,6))
plt.plot(x, 
         y, 
         color='steelblue', 
         linestyle='-', 
         marker='o',
        )

plt.title("Anotacion Bsica")
plt.xlabel("X")
plt.ylabel("Y")

props_flecha = {
    "arrowstyle" : "->",
    "connectionstyle" : "arc3, rad=0.3"
    }

plt.annotate("Punto destacado" , # el cuadro de text
             xy = (3,5), # donde apunta la flecha
             textcoords='offset points', # para que sea offset
             xytext=(-25, 40), # offset del punto sealado
                          # xytext = (2.75,6.25), # de donde quiero que salga (no offset)
             arrowprops= props_flecha); # propiedades de la flecha
```

```python

plt.plot(x, y, marker='o')
coordenadas = (3,5)
plt.annotate("Punto destacado", 
             xy = coordenadas, 
             textcoords='offset points', 
             xytext=(-25, 40),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0.2"))

plt.title('Anotacin Bsica')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

```python
plt.plot(x, y, marker='o')

# Aqu los annotate

coordenada1 = (1,2)
coordenada2 = (5,11)
plt.annotate("Inicio", 
             xy = coordenada1, 
             textcoords='offset points', 
             xytext=(100, 10),
             arrowprops=dict(arrowstyle="->",
                             connectionstyle="arc3",
                             color = "black",
                             linewidth = 5))

plt.annotate("Final", 
             xy = coordenada2, 
             textcoords='offset points', 
             xytext=(-50, -15),
             arrowprops=dict(arrowstyle="->",
                             connectionstyle="arc3",
                             color = "red",
                             linewidth = 5))


plt.title('Ejercicio 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Ejercicio 5

Carga el dataset de accidentes de trfico en EEUU que est en la ruta "./data/car_crashes.csv". Quedate con los cinco estados con menos accidentes en "total".

```python
df_cars = pd.read_csv("./data/car_crashes.csv", sep="|")
df_top_min = df_cars.nsmallest(5, 'total')
df_top_min
```

### Ejercicio 6

Completa el siguiente cdigo para obtener una figura como la representada en la imagen

```python
#OJO df_top_min contiene el subdataframe resultado del ejercicio anterior
estados = {
    "DC": "District of Columbia",
    "MA": "Massachusetts",
    "MN": "Minnesota",
    "WA": "Washington",
    "CT": "Connecticut"
}
for abrev,estado in estados.items():
    df_top_min["abbrev"] = df_top_min["abbrev"].str.replace(abrev,estado)
colores = ["green","gray","orange","blue","pink"]
plt.figure(figsize = (10,5))
plt.bar(df_top_min["abbrev"], height = df_top_min["total"], color = colores, alpha = 0.5)
plt.ylabel("Total accidentes por millones de millas recorridas")
plt.ylim(0,30)
plt.axhline(y = df_cars.total.max());
plt.title("Ejercicio 6");
```
