1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)

```python
# diagramas de barras

top_5 = df_cities.nlargest(5, 'population_total')
x = top_5["city"].to_list()
y = (top_5["population_total"]/1000).to_list()

colores = ["blue", "gray", "red" , "olive", "brown"]

for i in range(len(x)):
    plt.text( x[i], # x
             y[i]+100, # y
             f'{y[i]}', # texto
             ha = "center", 
             va = "bottom",
             fontsize = 7)

plt.bar(x, 
        y, 
        color= colores,
        alpha=0.5)
plt.title("Ciudades mas pobladas de California", fontsize = 7)
plt.ylabel("Habitantes (miles)", fontsize = 7)
plt.xlabel("Ciudades", fontsize = 7)
plt.yticks(range(0, 5100, 1000), fontsize = 7)
plt.xticks(fontsize = 7);
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)

```python
# histograma

plt.hist(df_cities["area_total_km2"],
         bins= 50,
         alpha = 0.8,   
         color = "steelblue",
         label = "Hist Area")

font_dict = {  #diccionario que usaremos
    "size" : 12,
    "family": "Bradley Hand itc",

}

plt.xlabel(r"Superficie $(Km^2)$", 
           fontdict = font_dict)
plt.ylabel("Numero de ciudades", 
           fontdict = font_dict)

plt.legend( fontsize = 7);
```

3. El scatter de rea de agua de un ciudad con su poblacion

```python
df_cities.loc[df_cities["area_water_km2"].idxmax(), "population_total"] /1000
```

```python
df_cities["area_water_km2"].max()
```

```python
df_cities["population_total"].max()/1000
```

```python
df_cities.loc[df_cities["population_total"].idxmax(), "area_water_km2"]
```

```python
x = df_cities["area_water_km2"]
y = df_cities["population_total"] / 1000

plt.scatter(x,
            y,
            s = 10,
            color = "g",
            alpha= 0.3)

outliers_1 = (df_cities.loc[df_cities["population_total"].idxmax(), "area_water_km2"],
              df_cities["population_total"].max()/1000) 

plt.annotate("Outlier",
             xy = outliers_1,
             xytext=(60,3500),
             arrowprops = dict(arrowstyle="->",
                             connectionstyle="arc3",
                             color = "black",
                             linewidth = 0.75)); # propiedades de la flecha

outliers_2 = (df_cities["area_water_km2"].max(),
              df_cities.loc[df_cities["area_water_km2"].idxmax(), "population_total"] /1000,
              )

plt.annotate("        ",
             xy = outliers_2,
             xytext=(60,3500),
             arrowprops = dict(arrowstyle="->",
                             connectionstyle="arc3",
                             color = "black",
                             linewidth = 0.75)); # propiedades de la flecha

plt.xlabel(r"Area de agua en la ciudad $(Km^2)$", fontsize = 7)
plt.ylabel("Poblacin (miles de habitantes)", fontsize = 7)
plt.yticks(fontsize = 7)
plt.xticks(fontsize = 7)
plt.title("Poblacion por kilmetro cuadrado de agua", fontsize = 7);
```

4. El boxplot de la altura de las ciudades

```python
plt.boxplot(df_cities["elevation_m"],
            whis = 1.5,
            vert = False,
            patch_artist = True, 
            boxprops = dict(color="black", linewidth=1, facecolor="red"),
            medianprops = dict(color="black", linewidth=1));

plt.xlabel("Altitud (m)",fontsize = 7)
plt.yticks(fontsize = 7)
plt.xticks(fontsize = 7)
plt.title("Diagrama de caja de la Altitud de las ciudades de California", fontsize = 7);
```

```python
# TODOS JUNTOS

# Figura y subgrficos: 2 filas x 2 columnas
fig, ax = plt.subplots(2, 2, figsize=(12, 12))
fig.subplots_adjust(hspace = 0.25, wspace = 0.25)

# Grfico 1: Diagrama de Barras
top_5 = df_cities.nlargest(5, 'population_total')
x = top_5["city"].to_list()
y = (top_5["population_total"]/1000).to_list()

colores = ["blue", "gray", "red" , "olive", "brown"]

for i in range(len(x)):
    ax[0, 0].text(  x[i], # x
                    y[i]+100, # y
                    f'{y[i]}', # texto
                    horizontalalignment = "center", 
                    verticalalignment = "bottom",
                    fontsize = 7)

ax[0, 0].bar(x, 
        y, 
        color= colores,
        alpha=0.5,
        )
ax[0, 0].set_title("Ciudades mas pobladas de California", fontsize = 7)
ax[0, 0].set_ylabel("Habitantes (miles)", fontsize = 7)
ax[0, 0].set_xlabel("Ciudades", fontsize = 7)
ax[0, 0].set_yticks(range(0, 5100, 1000))
ax[0, 0].tick_params(axis='x', labelsize=7)
ax[0, 0].tick_params(axis='y', labelsize=7)

# Grfico 2: Histograma
ax[0, 1].hist(df_cities["area_total_km2"],
         bins= 50,
         alpha = 0.8,   
         color = "steelblue",
         label = "Hist Area")

font_dict = {  #diccionario que usaremos
    "size" : 9,
    "family": "Bradley Hand itc",}

ax[0, 1].set_xlabel(r"Superficie $(Km^2)$", 
           fontdict = font_dict)
ax[0, 1].set_ylabel("Numero de ciudades", 
           fontdict = font_dict)
ax[0, 1].legend( fontsize = 7)
ax[0, 1].tick_params(axis='x', labelsize=7)
ax[0, 1].tick_params(axis='y', labelsize=7)


# Grfico 3: Scatterplot
x = df_cities["area_water_km2"]
y = df_cities["population_total"] / 1000

ax[1, 0].scatter(x,
            y,
            s = 10,
            color = "g",
            alpha= 0.3)

outliers_1 = (df_cities.loc[df_cities["population_total"].idxmax(), "area_water_km2"],
              df_cities["population_total"].max()/1000) 

ax[1, 0].annotate("Outlier",
             xy = outliers_1,
             xytext=(60,3500),
             fontsize = 7,
             arrowprops = dict(arrowstyle="->",
                             connectionstyle="arc3",
                             color = "black",
                             linewidth = 0.5)); # propiedades de la flecha

outliers_2 = (df_cities["area_water_km2"].max(),
              df_cities.loc[df_cities["area_water_km2"].idxmax(), "population_total"] /1000,
              )

ax[1, 0].annotate("        ",
             xy = outliers_2,
             xytext=(60,3500),
             arrowprops = dict(arrowstyle="->",
                             connectionstyle="arc3",
                             color = "black",
                             linewidth = 0.75)); # propiedades de la flecha

ax[1, 0].set_xlabel(r"Area de agua en la ciudad $(Km^2)$", fontsize = 7)
ax[1, 0].set_ylabel("Poblacin (miles de habitantes)", fontsize = 7)
ax[1, 0].tick_params(axis='x', labelsize=7)
ax[1, 0].tick_params(axis='y', labelsize=7)
ax[1, 0].set_title("Poblacion por kilmetro cuadrado de agua", fontsize = 7)

# Grfico 4: Boxplot
ax[1, 1].boxplot(df_cities["elevation_m"],
            whis = 1.5,
            vert = False,
            patch_artist = True,
            flierprops= dict(markersize = 3), 
            boxprops = dict(color="black", linewidth=0.5, facecolor="red"),
            medianprops = dict(color="black", linewidth=0.5))

ax[1, 1].set_xlabel("Altitud (m)", fontsize = 7)
ax[1, 1].set_title("Diagrama de caja de la Altitud de las ciudades de California", fontsize = 7)
ax[1, 1].tick_params(axis='x', labelsize=7)
ax[1, 1].tick_params(axis='y', labelsize=7);
```
